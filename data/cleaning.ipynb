```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('raw/movies.csv')
```

```python
data
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data.drop('total_reviews', axis=1, inplace=True)
data.drop('total_ratings', axis=1, inplace=True)
data.drop('release_date_(theaters)', axis=1, inplace=True)
data.drop('release_date_(streaming)', axis=1, inplace=True)
data.drop('box_office_(gross_usa)', axis=1, inplace=True)
data.drop('runtime', axis=1, inplace=True)
data.drop('production_co', axis=1, inplace=True)
data.drop('sound_mix', axis=1, inplace=True)
data.drop('aspect_ratio', axis=1, inplace=True)
data.drop('view_the_collection', axis=1, inplace=True)
data.drop('link', axis=1, inplace=True)
```

```python
data
```

```python
null_values = data.isnull().sum()
null_values
```

```python
# Eliminar filas donde la columna 'consensus' es nula
data = data.dropna(subset=['consensus'])

# Imputar valores faltantes en 'rating' con el valor más común
most_common_rating = data['rating'].mode()[0]
data.loc[:, 'rating'] = data['rating'].fillna(most_common_rating)

# Eliminar filas donde la columna 'genre' es nula
data = data.dropna(subset=['genre'])

# Imputar valores faltantes en 'original_language' con "unknown"
data.loc[:, 'original_language'] = data['original_language'].fillna('unknown')

# Eliminar fila donde la columna 'director' es nula
data = data.dropna(subset=['director'])

# Imputar valores faltantes en 'producer' y 'writer' con "unknown"
data.loc[:, 'producer'] = data['producer'].fillna('unknown')
data.loc[:, 'writer'] = data['writer'].fillna('unknown')

# Eliminar filas donde la columna 'synopsis' es nula
data = data.dropna(subset=['synopsis'])

# Imputar valores faltantes en 'people_score' con la media
mean_people_score = data['people_score'].mean()
data.loc[:, 'people_score'] = data['people_score'].fillna(mean_people_score)
```

```python
null_values = data.isnull().sum()
null_values
```

```python
data.to_csv('processed/movies_cleaned.csv', index=False)
```
